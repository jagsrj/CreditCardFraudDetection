```python
df_card_transactions['transaction_dt2']=pd.to_datetime(df_card_transactions['transaction_dt'])
```

```python
# Sort descending order
df_card_transactions.sort_values(by=['member_id','transaction_dt2'], inplace=True,
               ascending = [True, False])
print(df_card_transactions)
```

```python
df_card_transactions["Rank"] = df_card_transactions.groupby("member_id")["transaction_dt2"].rank(ascending=False).astype(int)
df_card_transactions.sort_values("Rank")
```

```python
#df_card_transactions_top10 = df_card_transactions[(df_card_transactions.Rank <=10) & (df_card_transactions.status.str.contains('GENUINE'))]
df_card_transactions_top10 = df_card_transactions[(df_card_transactions.Rank <=10) ]
df_card_transactions_top10
```

```python
#df2 = df.groupby('Courses')['Fee'].agg(['sum','count'])
df2 = df_card_transactions_top10.groupby(['member_id'])['amount'].agg(['mean', 'std'])
#df2 = df_card_transactions_top10.groupby(['card_id'])['amount'].agg(['mean', 'std'])
print(df2)
```

```python
df2 = df2.reset_index()
df2
```

```python
df2.dtypes
```

```python
df2.rename(columns = {'mean':'Avg1', 'std':'SD'}, inplace = True)
```

```python
df2['ucl'] = df2.Avg1 +  3*df2.SD
```

```python
df2
```

```python
lstTxnPostCodeTxDate = df_card_transactions[df_card_transactions.Rank <=1]
lstTxnPostCodeTxDate
```

```python
tmp1 = lstTxnPostCodeTxDate.merge(df2, on='member_id', how='left')
tmp2 = tmp1.merge(df_member_score, on='member_id', how='left')
tmp2
```

```python
columns = ['amount', 'pos_id', 'status', 'transaction_dt', 'status', 'Rank', 'Avg1', 'SD']
tmp2.drop(columns, inplace=True, axis=1)
```

```python
print(tmp2)
```

```python
lookup_df = tmp2
lookup_df
```
